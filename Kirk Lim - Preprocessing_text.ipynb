```python
import pandas as pd
import EDAHelper as hlpr
```

```python
start_time = hlpr.show_time()
print("Start of the run ...",start_time)
```

```python
df = pd.read_csv('kdata.csv')
#df = pd.read_csv('sample.csv')
```

```python
df
```

```python
df['new_text']=df['text']
```

```python
print(df['new_text'].dtype)
```

```python
df.info()
```

# LOWER CASE 

```python
df['new_text']=df['new_text'].apply(hlpr.lower_cases)
```

```python
df
```

# REMOVE HTML TAGS

```python
df['new_text']=df['new_text'].apply(hlpr.remove_html_tags)
```

```python
df
```

# REMOVE ANY URL

```python
df['new_text']=df['new_text'].apply(hlpr.remove_url)
```

```python
df
```

# REMOVE PUNCTUATION MARKS

```python
df['new_text']=df['new_text'].apply(hlpr.remove_punc)
```

```python
df
```

# REMOVE STOPWORDS

```python
df['new_text']=df['new_text'].apply(hlpr.remove_stopwords)
```

```python
df
```

# REMOVE EMOJIS

```python
df['new_text']=df['new_text'].apply(hlpr.remove_emojis_manually)
```

```python
df
```

```python
df.to_csv('kdata_clean.csv')
```

# REPLACE CHAT ABBREVIATION WITH WORDS / CHAT WORDS BASED ON https://www.kaggle.com/code/nmaguette/up-to-date-list-of-slangs-for-text-preprocessing/notebook

```python
df['new_text']=df['new_text'].apply(hlpr.chat_conversion)
```
